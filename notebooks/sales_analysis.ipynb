```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
sales_category = df.groupby("Category")["Sales"].sum().sort_values()

plt.figure(figsize=(8,5))
sales_category.plot(kind="bar")

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
profit_category = df.groupby("Category")["Profit"].sum().sort_values()

plt.figure(figsize=(8,5))
profit_category.plot(kind="bar")

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.show()
```

```python
sales_region = df.groupby("Region")["Sales"].sum().sort_values()

plt.figure(figsize=(8,5))
sales_region.plot(kind="bar")

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

```python
monthly_sales = df.groupby("Month Number")["Sales"].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.xticks(range(1,13))

plt.show()
```

```python
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_products.plot(kind="bar")

plt.title("Top 10 Products by Sales")
plt.ylabel("Sales")

plt.show()
```

```python
top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_customers.plot(kind="bar")

plt.title("Top 10 Customers")
plt.ylabel("Sales")

plt.show()
```

```python
numeric_df = df[["Sales","Quantity","Discount","Profit"]]

plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

# Business Insights

# Business Insights

## 1. Sales Performance
- Technology generated the highest sales among all categories.
- Office Supplies maintained consistent sales across the dataset.
- Furniture contributed the least sales compared to the other categories.

## 2. Profit Analysis
- Technology generated the highest profit.
- Some Furniture products had lower profitability despite good sales.
- Higher sales do not always guarantee higher profits.

## 3. Regional Analysis
- Sales performance varies across different regions.
- Some regions contribute significantly more revenue than others, indicating opportunities for regional marketing strategies.

## 4. Customer Analysis
- A small number of customers account for a large portion of total sales.
- These high-value customers are important for customer retention strategies.

## 5. Product Analysis
- Certain products generate high revenue and should always remain in stock.
- Products with low profitability should be reviewed for pricing and discount policies.

## 6. Discount Analysis
- The correlation between Discount and Profit is negative (-0.22).
- Increasing discounts generally reduces overall profit.

## 7. Correlation Analysis
- Sales and Profit show a moderate positive correlation (0.48).
- Quantity has only a weak relationship with Profit.
- Discount has a negative impact on Profit.
